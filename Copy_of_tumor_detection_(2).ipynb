```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

## **loading the data**

```python
data = pd.read_csv("Tumor Cancer Prediction_Data.csv")
```

```python
data.head()
```

# **Data preproceesing**

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data=data.convert_dtypes()
```

```python
data.shape
```

```python
data.dtypes
```

```python
data["diagnosis"] = data["diagnosis"].replace(['B','M'],[0,1])
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.corr()
```

```python
data=data.drop(['Index','F5','F15','F20','F23'],axis=1)
```

```python
data.corr()
```

```python
X = data.drop(columns=['diagnosis'],axis=1) #features detection
Y = data['diagnosis'] #labeling
```

```python
# set aside 20% of train and test data for evaluation
X_train, X_test, y_train, y_test = train_test_split(X, Y,
    test_size=0.25, shuffle = True, random_state = 42)
```

```python
# Use the same function above for the validation set
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, 
    test_size=0.5, random_state= 42) 
```

```python
X_test.info()
```

```python
X_val.info()
```

# **Logistic Regression**

```python
# apply logistic regression
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression(solver='lbfgs',max_iter = 3000)
```

```python
#veiwing the score of data cross validation
from sklearn.model_selection import cross_val_score
cross_val = cross_val_score(Lr,X_val,y_val,cv = 5).mean()
print("*Cross Validation Score* ===> ",cross_val)
```

```python
Lr.fit(X, Y)
```

```python
#Accuracy of taraining data
y_pred = Lr.predict(X_train)
print(accuracy_score(y_train, y_pred))
```

```python
#Accuracy of testing data
y_pred = Lr.predict(X_test)
print(accuracy_score(y_test, y_pred))
```
